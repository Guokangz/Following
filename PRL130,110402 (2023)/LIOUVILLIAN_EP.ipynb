```python
import numpy as np
import cmath
from scipy.linalg import eigh
import matplotlib.pyplot as plt

r_eff = np.linspace(0, 8, 100)
omiga = np.ones(100, dtype=complex)
print(omiga)
S1 = 0.25 * (-3*r_eff - np.sqrt(np.power(r_eff,2) - 16*np.power(omiga,2)))
S2 = 0.25 * (-3*r_eff + np.sqrt(np.power(r_eff,2) - 16*np.power(omiga,2)))

engry_1 = np.imag(S1)
engry_2= np.imag(S2)

decay_1=np.real(S1)
decay_2=np.real(S2)


plt.figure()
# 绘制曲线
plt.plot(r_eff/4*omiga, engry_1, label='lamda3')
plt.plot(r_eff/4*omiga,engry_2, label='lamda4')

# 添加标题和标签
plt.title('Engry vs r_eff')
plt.xlabel('r_eff')
plt.ylabel('Engry')

# 添加图例
plt.legend()

# 显示图形
plt.show()

```

```python
plt.figure()

plt.plot(r_eff/4*omiga, decay_1, label='lamda3')
plt.plot(r_eff/4*omiga, decay_2, label='lamda4')
# 添加标题和标签
plt.title('Decay vs r_eff')
plt.xlabel('r_eff')
plt.ylabel('Decay')

# 添加图例
plt.legend()

# 显示图形
plt.show()
```
