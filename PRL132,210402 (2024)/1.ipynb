```python
import numpy as np
import matplotlib.pyplot as plt

## 主要参数
omega = 1
Gamma = 0.04*omega
kesai = 0.1*omega
kappa = 0.003*omega
Lambda = Gamma + kappa


def E_b(t):

    return (16*omega*Gamma**2*kesai**2/Lambda**4)*(1+np.exp(-Lambda*t/2)*(np.exp(-Lambda*t/2)-2))

def E_a(t):

    return (4*omega*kesai**2*(1-np.exp(-Lambda*t/2))**2/Lambda**2)


J = Gamma/2



E_a_dot = []
E_b_dot = []
E_p_dot = []
N =500
tstep = J / N
for i in range(0, N):

    t = i * tstep
    E_b_dot.append(E_b(i))
    E_a_dot.append(E_a(i))
    E_p_dot.append(E_b(i) / E_a(i))


t = np.linspace(0, J*500, N)
fig = plt.figure(figsize=(12, 5))
plt.plot(t, E_a_dot, label='E_a')
plt.plot(t, E_b_dot, label='E_b')
plt.legend()

```
